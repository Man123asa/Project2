```python
## Grouping of data based on the Region , Country and Item type of the order

d1=data.groupby(['Region','Country','Item Type'])
d1.first()
```

```python
## Dataset containing information: Total revenue, Total cost, Total profit

f3=data[["Region","Total Revenue","Total Cost","Total Profit"]]

data1=f3.groupby('Region')['Total Revenue'].apply(lambda x:x.values.sum())
data2=f3.groupby('Region')['Total Cost'].apply(lambda y:y.values.sum())
data3=f3.groupby('Region')['Total Profit'].apply(lambda x:x.values.sum())

print(data1,"\n\n",data2,"\n\n",data3)
```

```python
import matplotlib.pyplot as pt
import numpy as n

## Set height of the bar
a=n.array(data1)
b=n.array(data2)
c=n.array(data3)

## set width of the bar
width=0.25
fig=pt.subplots(figsize=(20,10))

## Set position of the bar on x axis    
br1=n.arange(len(a))
br2=[x+width for x in br1]
br3=[x+width for x in br2]

## Set the plot
pt.bar(br1,a,color='olive',width=width,edgecolor='k',label='Total revenue')
pt.bar(br2,b,color='mediumseagreen',width=width,edgecolor='k',label='Total cost')
pt.bar(br3,c,color='aquamarine',width=width,edgecolor='k',label='Total profit')

## Setting the x ticks and x,y labels
pt.xlabel("Region",color='black',family='cursive',fontsize=25)
pt.ylabel("Revenue flow",color='Black',family='cursive',fontsize=25)
pt.xticks([r+width for r in range(len(a))],['Asia','Australia & Oceania','Central America & The Carrebian','Europe','Middle East and North Africa','North America','Sub-Sahara Africa'])

pt.title('Profit per Cost in different regions',family='cursive',fontsize=30) 
pt.legend()
pt.show()
```

```python
## Dataset to find the major selling products
## This dataset includes the information: Item Tyoe and the units sold per item type
## Here used sorting method Loc to get the desired data using the column names

d=data.loc[:,["Item Type","Units Sold"]]

## Grouping the dataset based on the item type
## Used the groupby function to form the group with respect to "Item type"

hello=d.groupby('Item Type').apply(lambda x:x.values.sum())
print(hello)
```

```python
import matplotlib.pyplot as p
import numpy as n

## Setting the values and the labels to be displayed 
y=n.array(hello)
l=["Baby food","Beverages","Cereal","Clothes","Cosmetics","Fruits","Household","Meat","Office supplies","Personal care","Snacks","Vegetables"]

## Using explode to highlight the most and least sold goods
ex=[0,0,0,0,0.3,0,0,0.4,0,0,0,0]
## Setting the pie chart
p.pie(y,labels=l,explode=ex,radius=1,shadow=True)
font2={'family':'monospace','color':'k','size':20}
p.title("MOST SOLD ITEMS",fontweight='light',fontdict=font2,loc="center")
p.show()
```
